Using your Scrabble `Word` classes created in previous assignment, create a `Word` object for each allowed word in the `scrabble_list.txt` file. Add each instance of you `Word` object to a list and then pickle that list to a file named `scrabble_list.pickle`.

One of the advantages of Pickling is that the binary data format is supposed to be faster to read in and process. Compare the time to read in the text file and convert to `Word` objects versus reading in the pickled file. 



```python
import importlib
import scrabble

importlib.reload(scrabble)
from scrabble import Word
import re
import pickle
import time

SCRABBLE_WORDS_FILE = "scrabble_list.txt"
WORD_EXTRACTOR_PATTERN = re.compile(r"(?:^|\n)([a-zA-z]+)(?:\n|$)")

all_words = WORD_EXTRACTOR_PATTERN.findall(open(SCRABBLE_WORDS_FILE).read())
words_and_scores = []
start = time.time()
for word in all_words:
    words_and_scores.append(Word(word))
end = time.time()
file_time = end - start
print(f"Text file time: {file_time:.4f} seconds")
pickle_filename = "scrabble_list.pickle"

with open(pickle_filename, "wb") as file:
    pickle.dump(words_and_scores, file)
```
